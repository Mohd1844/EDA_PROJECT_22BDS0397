------DATA CLEANING AND PREPROCESSING------

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df['Region'] = df['Region'].astype('category')
df['Country'] = df['Country'].astype('category')
df['Item Type'] = df['Item Type'].astype('category')
df['Sales Channel'] = df['Sales Channel'].astype('category')
df['Order Priority'] = df['Order Priority'].astype('category')
```

```python
#Check Data Type
df.dtypes
```

```python
#Checking Missing Value
df.isnull().sum()
```

```python
#Checking for the duplicate values
df.duplicated().sum()
```

```python
#  Drop exact duplicate rows
initial_rows = df.shape[0]
df = df.drop_duplicates()
print(f"Dropped {initial_rows - df.shape[0]} duplicate rows.")
```

```python
# Drop rows where Order Date is missing
df = df.dropna(subset=['Order Date'])


```

```python

if 'Order Date' in df.columns and 'Ship Date' in df.columns:
    df['OrderYear'] = df['Order Date'].dt.year
    df['OrderMonth'] = df['Order Date'].dt.month
    df['OrderMonthName'] = df['Order Date'].dt.strftime('%b')
    df['OrderWeekday'] = df['Order Date'].dt.day_name()
    df['LeadTimeDays'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
num_cols = ['Units Sold','Unit Price','Unit Cost','Total Revenue','Total Cost','Total Profit']
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```

## ---EXPLORATORY DATA ANALYSIS AND VISUALIZATION---

```python
# ---  Univariate analysis ---
```

```python
numeric = df.select_dtypes(include=[np.number]).columns.tolist()
print('\nNumeric columns:', numeric)
```

```python
for col in numeric:
    plt.figure()
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'22BDS0397  Distribution of {col}')
    plt.show()
```

```python
# Categorical counts
cat_candidates = df.select_dtypes(include=['category','object']).columns.tolist()
print('\nCategorical columns:', cat_candidates)
```

```python
for col in cat_candidates:
    plt.figure()
    order = df[col].value_counts().index
    sns.countplot(y=col, data=df, order=order)
    plt.title(f'22BDS0397 Countplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# --- Bivariate analysis ---
```

```python
correlation = df['Total Revenue'].corr(df['Total Profit'])
print(f"Correlation between Total Revenue and Total Profit: {correlation:.3f}")
```

```python
if 'Units Sold' in df.columns and 'Total Profit' in df.columns:
    plt.figure()
    sns.scatterplot(x='Units Sold', y='Total Profit', data=df)
    plt.title('22BDS0397 Units Sold vs Total Profit')
    plt.show()
```

This scatter plot shows a strong positive correlation between Units Sold and Total Profit. The data points form several distinct lines, suggesting that different product categories exist, each with a different, consistent profit margin.

```python
# Price vs Profit per unit
if 'Unit Price' in df.columns and 'Unit Cost' in df.columns:
    df['ProfitPerUnit'] = df['Unit Price'] - df['Unit Cost']
    plt.figure()
    sns.scatterplot(x='Unit Price', y='ProfitPerUnit', data=df)
    plt.title('22BDS0397 Unit Price vs Profit per Unit')
    plt.show()
```

This scatter plot shows a positive correlation between Unit Price and Profit per Unit. As the unit price increases, the profit per unit generally increases as well. The relationship isn't perfectly linear, but a clear upward trend is visible.

```python
# Categorical vs numeric: boxplots
for cat in ['Region','Item Type','Sales Channel','Order Priority']:
    if cat in df.columns and 'Total Profit' in df.columns:
        plt.figure(figsize=(12,6))
        sns.boxplot(x=cat, y='Total Profit', data=df)
        plt.title(f'22BDS0397 Total Profit by {cat}')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Correlation heatmap for numeric columns
if len(numeric) >= 2:
    plt.figure(figsize=(10,8))
    corr = df[numeric].corr()
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', center=0)
    plt.title('22BDS0397 Correlation matrix')
    plt.show()
```

Financial metrics like Unit Price, Unit Cost, Total Revenue, Total Cost, and Total Profit are all strongly positively correlated with each other.

Units Sold has a moderate positive correlation with these financial metrics.

Time-based data (OrderYear, OrderMonth, LeadTimeDays) and Order ID have almost no correlation with the financial performance.

```python
# ---  Multivariate analysis ---
# Pivot: Total Profit by Region and Year
if 'Total Profit' in df.columns and 'Region' in df.columns and 'OrderYear' in df.columns:
    pivot = df.pivot_table(index='Region', columns='OrderYear', values='Total Profit', aggfunc='sum')
    print('\nTotal Profit by Region and Year:')
    display(pivot)
    pivot.plot(kind='bar', stacked=False)
    plt.title('22BDS0397 Total Profit by Region and Year')
    plt.ylabel('Total Profit')
    plt.show()
```

Top Performers: Europe and Sub-Saharan Africa are the most profitable regions.

Performance Peaks:

Europe saw an exceptionally high profit in 2015.

Sub-Saharan Africa had its best years in 2012 and 2014.

Lowest Performer: North America consistently generated the lowest profit among all regions shown.

Inconsistent Trends: There is no consistent year-over-year growth; profits fluctuate significantly within each region annually.

```python
# Pairplot for selected numeric columns


pair_cols = [c for c in numeric if c in ['Units Sold','Unit Price','Unit Cost','Total Profit','Total Revenue']]
if len(pair_cols) >= 2 and len(df) <= 1000:
    sns.pairplot(df[pair_cols].dropna())
    plt.show()
```

```python
# Time series: monthly revenue or profit trends
if 'Order Date' in df.columns and 'Total Revenue' in df.columns:
    ts = df.set_index('Order Date').resample('M')['Total Revenue'].sum()
    plt.figure()
    ts.plot(marker='o')
    plt.title('22BDS03397 Monthly Total Revenue')
    plt.ylabel('Total Revenue')
    plt.show()
```

High Volatility: The revenue fluctuates significantly from month to month, showing many sharp peaks and troughs.

Seasonality: The repeating pattern of ups and downs suggests a strong seasonal effect on revenue.

No Clear Trend: There isn't a clear long-term upward or downward trend over the years; the revenue oscillates around an average level.

Notable Peak: The highest revenue peak occurred in early 2013.

```python
if 'Order Date' in df.columns and 'Total Profit' in df.columns:
    ts2 = df.set_index('Order Date').resample('M')['Total Profit'].sum()
    plt.figure()
    ts2.plot(marker='o')
    plt.title('22BDS0397 Monthly Total Profit')
    plt.ylabel('Total Profit')
    plt.show()
```

The chart shows that profit is highly volatile and seasonal, with regular peaks and troughs each year. There is no clear long-term upward or downward trend. The highest profit peak occurred in early 2013.
